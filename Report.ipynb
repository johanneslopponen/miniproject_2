# Training the digits

Computing the SVD for 10 variants of each digit.

```python
TrainDigits = np.load("HandwrittenDigits/TrainDigits.npy")
TrainLabels = np.load("HandwrittenDigits/TrainLabels.npy")

u_dict = {i: [] for i in range(10)}

for i in range(110):
    d = TrainDigits[:,i] # The first digit in the training set
    D = np.reshape(d, (28, 28)).T # Reshaping a vector to a matrix
    label = TrainLabels[:,i].item()
    U, S, V = np.linalg.svd(D)

    u_dict[label].append(U)

for label in u_dict:
    u_dict[label] = np.array(u_dict[label])

# Tesing
print(len(u_dict[0]))

def residual(digit, U_k):
    res = np.linalg.norm((np.eye(len(U_k)) - U_k@U_k.T)@digit,2)
    return res
```
